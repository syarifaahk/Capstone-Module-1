```python
# Data dummy untuk ruangan penyimpanan
storage_rooms = {
    'Room A': {'capacity': 100, 'current_stock': 80, 'temperature': 2},
    'Room B': {'capacity': 50, 'current_stock': 40, 'temperature': 5},
    'Room C': {'capacity': 150, 'current_stock': 120, 'temperature': -20},
}

# Data vaksin
vaccines = [
    {'name': 'Vaccine A', 'batch': '001A', 'expiry': '2025-01-01', 'stock': 20, 'temperature': 2, 'room': 'Room A'},
    {'name': 'Vaccine B', 'batch': '002B', 'expiry': '2024-10-15', 'stock': 10, 'temperature': 5, 'room': 'Room B'},
    {'name': 'Vaccine C', 'batch': '003C', 'expiry': '2026-03-22', 'stock': 30, 'temperature': -20, 'room': 'Room C'},
]

def display_menu():
    print("1. Input Vaksin Baru (Create)")
    print("2. Tampilkan Stok Vaksin (Read)")
    print("3. Keluar")

def create_vaccine():
    name = input("Masukkan nama vaksin: ")
    batch = input("Masukkan nomor batch: ")
    expiry = input("Masukkan tanggal kedaluwarsa (YYYY-MM-DD): ")
    stock = int(input("Masukkan jumlah stok: "))
    temperature = int(input("Masukkan suhu penyimpanan: "))
    
    # Cek apakah ada ruangan dengan suhu yang sesuai
    suitable_room = None
    for room, info in storage_rooms.items():
        if info['temperature'] == temperature:
            suitable_room = room
            break
    
    if not suitable_room:
        print("Tidak ada ruangan dengan suhu yang sesuai. Stok baru tidak bisa ditambahkan.")
        return
    
    # Cek kapasitas ruangan
    if storage_rooms[suitable_room]['current_stock'] + stock > storage_rooms[suitable_room]['capacity']:
        print(f"Ruangan {suitable_room} penuh. Tidak bisa menambah stok.")
        return
    
    # Tambahkan vaksin ke data
    vaccines.append({'name': name, 'batch': batch, 'expiry': expiry, 'stock': stock, 'temperature': temperature, 'room': suitable_room})
    storage_rooms[suitable_room]['current_stock'] += stock
    print(f"Vaksin {name} ditambahkan ke {suitable_room} dengan suhu {temperature}°C.")

def read_vaccine_stock():
    for v in vaccines:
        print(f"Vaksin: {v['name']}, Batch: {v['batch']}, Expiry: {v['expiry']}, Stok: {v['stock']}, Suhu: {v['temperature']}°C, Ruang: {v['room']}")
    
    print("1. Update Stok")
    print("2. Hapus Stok")
    choice = int(input("Pilih opsi: "))
    
    if choice == 1:
        update_stock()
    elif choice == 2:
        remove_stock()

def update_stock():
    batch = input("Masukkan nomor batch vaksin yang ingin diperbarui: ")
    for v in vaccines:
        if v['batch'] == batch:
            room = storage_rooms[v['room']]
            additional_stock = int(input("Masukkan jumlah stok yang ingin ditambahkan: "))
            if room['current_stock'] + additional_stock > room['capacity']:
                print(f"Tidak bisa update stok vaksin {v['name']} karena ruangan penuh!")
            else:
                v['stock'] += additional_stock
                room['current_stock'] += additional_stock
                print(f"Stok vaksin {v['name']} berhasil diperbarui!")
            return
    print("Batch vaksin tidak ditemukan!")

def remove_stock():
    batch = input("Masukkan nomor batch vaksin yang ingin dihapus: ")
    for v in vaccines:
        if v['batch'] == batch:
            storage_rooms[v['room']]['current_stock'] -= v['stock']
            vaccines.remove(v)
            print(f"Vaksin dengan batch {batch} berhasil dihapus!")
            return
    print("Batch vaksin tidak ditemukan!")

# Main loop
while True:
    display_menu()
    choice = int(input("Pilih menu: "))
    
    if choice == 1:
        create_vaccine()
    elif choice == 2:
        read_vaccine_stock()
    elif choice == 3:
        print("Terima kasih telah menggunakan aplikasi manajemen vaksin!")
        break
    else:
        print("Pilihan tidak valid!")
```

```python
# Dummy Data Vaksin dan Storage
vaccines = [
    {"name": "Vaksin A", "batch_number": "A001", "expiry_date": "2025-05-01", "stock": 100, "required_temp": 2},
    {"name": "Vaksin B", "batch_number": "B001", "expiry_date": "2024-10-01", "stock": 50, "required_temp": -20}
]

storage_rooms = [
    {"room_id": 1, "temperature": 2, "capacity": 200, "used_capacity": 100},  # Menampung Vaksin A
    {"room_id": 2, "temperature": -20, "capacity": 100, "used_capacity": 50},  # Menampung Vaksin B
]

# Fungsi utama aplikasi manajemen stok vaksin
def main():
    while True:
        print("\n--- Main Menu ---")
        print("1. Create (Input Vaksin Baru)")
        print("2. Read (Melihat Stok Vaksin)")
        print("3. Exit")
        choice = input("Pilih opsi: ")

        if choice == "1":
            create_vaccine()
        elif choice == "2":
            read_vaccines()
        elif choice == "3":
            print("Selesai.")
            break
        else:
            print("Opsi tidak valid, coba lagi.")

# Fungsi untuk input vaksin baru
def create_vaccine():
    print("\n--- Input Vaksin Baru ---")
    name = input("Nama Vaksin: ")
    batch_number = input("Nomor Batch: ")
    expiry_date = input("Tanggal Kedaluwarsa (YYYY-MM-DD): ")
    stock = int(input("Jumlah Stok: "))
    required_temp = int(input("Suhu Penyimpanan yang Dibutuhkan: "))

    # Cek ruangan yang cocok
    suitable_room = None
    for room in storage_rooms:
        if room["temperature"] == required_temp:
            suitable_room = room
            break

    if not suitable_room:
        print("Tidak ada ruangan dengan suhu yang sesuai. Stok baru tidak bisa ditambahkan.")
        return

    # Cek kapasitas ruangan
    available_capacity = suitable_room["capacity"] - suitable_room["used_capacity"]
    if suitable_room["used_capacity"] + stock > suitable_room["capacity"]:
        print(f"Ruangan penyimpanan penuh. Kapasitas hanya tersedia: {available_capacity} unit.")
    else:
        # Tambahkan vaksin ke dalam stok
        vaccines.append({
            "name": name,
            "batch_number": batch_number,
            "expiry_date": expiry_date,
            "stock": stock,
            "required_temp": required_temp
        })
        suitable_room["used_capacity"] += stock
        print("Vaksin berhasil ditambahkan ke stok.")

# Fungsi untuk melihat stok vaksin
def read_vaccines():
    print("\n--- Stok Vaksin ---")
    if not vaccines:
        print("Tidak ada stok vaksin yang tersedia.")
        return

    for vaccine in vaccines:
        print(f"Nama: {vaccine['name']}, Batch: {vaccine['batch_number']}, Kedaluwarsa: {vaccine['expiry_date']}, Stok: {vaccine['stock']}, Suhu: {vaccine['required_temp']}°C")
    
    # Pilihan untuk update atau remove
    print("\nPilih opsi:")
    print("1. Update Stok")
    print("2. Remove Stok")
    choice = input("Pilih opsi: ")

    if choice == "1":
        update_stock()
    elif choice == "2":
        remove_stock()
    else:
        print("Opsi tidak valid.")

# Fungsi untuk update stok vaksin
def update_stock():
    batch_number = input("\nMasukkan nomor batch vaksin yang ingin di-update: ")
    for vaccine in vaccines:
        if vaccine["batch_number"] == batch_number:
            # Cari ruangan yang sesuai
            for room in storage_rooms:
                if room["temperature"] == vaccine["required_temp"]:
                    available_capacity = room["capacity"] - room["used_capacity"]
                    print(f"Stok saat ini: {vaccine['stock']}")
                    print(f"Kapasitas ruangan tersisa: {available_capacity} unit")

                    update_choice = input("Apakah ingin menambah atau mengurangi stok? (tambah/kurang): ").lower()
                    if update_choice not in ["tambah", "kurang"]:
                        print("Opsi tidak valid.")
                        return

                    update_amount = int(input("Jumlah stok yang ingin di-update: "))
                    
                    if update_choice == "tambah":
                        if room["used_capacity"] + update_amount > room["capacity"]:
                            print(f"Ruangan penuh! Kapasitas hanya tersedia: {available_capacity} unit. Stok tidak bisa di-update.")
                            return
                        else:
                            vaccine["stock"] += update_amount
                            room["used_capacity"] += update_amount
                            print("Stok berhasil ditambahkan.")
                            return
                    
                    elif update_choice == "kurang":
                        if update_amount > vaccine["stock"]:
                            print("Jumlah pengurangan lebih besar dari stok yang ada. Stok tidak bisa di-update.")
                            return
                        else:
                            vaccine["stock"] -= update_amount
                            room["used_capacity"] -= update_amount
                            print("Stok berhasil dikurangi.")
                            return

    print("Batch number tidak ditemukan.")

# Fungsi untuk menghapus vaksin
def remove_stock():
    batch_number = input("\nMasukkan nomor batch vaksin yang ingin dihapus: ")
    for vaccine in vaccines:
        if vaccine["batch_number"] == batch_number:
            # Kembalikan kapasitas ruangan
            for room in storage_rooms:
                if room["temperature"] == vaccine["required_temp"]:
                    room["used_capacity"] -= vaccine["stock"]
            vaccines.remove(vaccine)
            print("Vaksin berhasil dihapus.")
            return
    print("Batch number tidak ditemukan.")

# Mulai aplikasi
main()
```
